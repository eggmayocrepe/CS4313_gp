```python
'''
Author search - current functions
1) input case corrected
2) can search for first name (gerard), last name (goggin), full name (gerard goggin) or partial name (gog)
3) repeated results within the same paper is not repeated in search (e.g. if you search 'g', you won't get the same paper twice)

'''

import json

f1 = open('master.json', encoding='utf-8')   

v1 = json.loads(f1.read())     

i = 0

author_search = input("Search for author: ").lower()

for papers in v1:
    author_list = []   
    for author in range (len(papers["authors"])):
        fullname = papers["authors"][author]["fname"] +" "+ papers["authors"][author]["lname"]
        author_list.append(fullname)
    for fullname in author_list:
        if author_search in fullname.lower():
            print()
            print (f"Paper ID: {papers['paper_id']}\nTitle: {papers['paper']['title']}\nAuthors:")
            print(*author_list, sep = ', ')
#            print(author_list)            this version prints in a list, sucks
            i = i + 1
            break
        
print (f"\nAuthor with search term [{author_search.title()}], published {i} articles.")
            
f1.close()
```

```python
#hongern's limit code
import json

f1 = open('dataset.json', encoding='utf-8')   

v1 = json.loads(f1.read())      

count = 0
for papers in v1:
    
    all_authors = []
    for authors in papers["authors"]:
        all_authors.append(authors["fname"])
    print(f'{papers["paper_id"]}:{all_authors}') 
    
    count = count + 1
    if count >= 10:
        flag = "Red"
        while flag == "Red":

            view = input(f"Type Y/N to view more: ")
            if view == "Y" or view == "y":
                flag = "Y"
                count = 0
                break

            elif view == "N" or view == "n":
                flag = "N"
                break

            else:
                print("You have entered an unexpected input.")

        if flag == "N":
            break
    
f1.close()   
```
