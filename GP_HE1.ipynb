```python
# Retrieve fname of authors of every publications 
# Limit number of output
import json

f1 = open('dataset.json', encoding='utf-8')   

v1 = json.loads(f1.read())      

count = 0
for papers in v1:
    
    all_authors = []
    for authors in papers["authors"]:
        all_authors.append(authors["fname"])
    print(f'{papers["paper_id"]}:{all_authors}') 
    
    count = count + 1
    if count >= 10:
        flag = "Red"
        while flag == "Red":

            view = input(f"Type Y/N to view more: ")
            if view == "Y" or view == "y":
                flag = "Y"
                count = 0
                break

            elif view == "N" or view == "n":
                flag = "N"
                break

            else:
                print("You have entered an unexpected input.")

        if flag == "N":
            break
    
f1.close()   
```

```python
# Search papers according to author name
import json

f1 = open('dataset.json', encoding='utf-8')   

v1 = json.loads(f1.read())      

while True:
    flag = "Red"
    author_name = input ("Please input the name of author for the publication you are looking for: ")
    author_name = author_name.lower()
    author_name_validity = str(author_name).replace(" ","").replace(".","").replace("-", "")   # to avoid numbers and symbols, except for "-" and " " .

    if author_name_validity.isalpha() == False:   
        print("\nYou have entered an unexpected input, please retry.")   # error msg incase user input invalid
        flag = "Green"
    
    else:
        for papers in v1:
            for count in range (len(papers["authors"])):
                aufirst = papers['authors'][count]['fname']
                aulast = papers['authors'][count]['lname']
                if author_name in aufirst.lower() or author_name in aulast.lower():
                    flag = "Green"
                    print(f"\nName: {aufirst} {aulast}\nPublication: {papers['paper']['title']}")
                
    if flag == "Red":
        print(f"\nAuthor name containing '{author_name}' not found, please retype.")
    
    

f1.close()
```

```python
# Search for papers co-written by two authors
import json

f1 = open("dataset.json", encoding = "utf-8")

v1 = json.loads(f1.read())

while True:
    print("\033[1mThis program helps to search for papers that are co-written by multiple authors.\033[0m")


    author = input("Please enter name of  first author: ")
    authornext = input("Please enter name of next author: ")
    
    author1 = author.lower()
    authornext1 = authornext.lower()

    fullname_list = []
    author_list = []
    fulllist = []
    paperlist = []
    capital_list = []
    
    
    for papers in v1:
        small_list = []
        mini_capital_list = []
        for allauthors in range(len(papers["authors"])):
            
            capitalname = papers["authors"][allauthors]["fname"] + " " + papers["authors"][allauthors]["lname"]
            mini_capital_list.append(capitalname)
        
            fullname = (papers["authors"][allauthors]["fname"]).lower() + " " + (papers["authors"][allauthors]["lname"]).lower()
            small_list.append(fullname)
        fulllist.append(small_list)
        capital_list.append(mini_capital_list)

#     print(fulllist)

    for papers in v1:
        paperlist.append(papers["paper"]["title"])
#     print(paperlist)
    
        
    for authors in range(len(fulllist)):
        flag1 = "outside"
        flag2 = "outside"
        for i in range(len(fulllist[authors])):
            if author1 in fulllist[authors][i]:
                    flag1 = "inside"

            if authornext1 in fulllist[authors][i]:
                    flag2 = "inside"
      
        if flag1 == "inside" and flag2 == "inside":
            print(f"\nPaper ID: {authors+1}\nPublication: {paperlist[authors]}\nAuthors: {capital_list[authors]}\n")


f1.close()
```
