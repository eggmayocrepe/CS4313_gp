```python
# Retrieve fname of authors of every publications 
import json

f1 = open('dataset.json', encoding='utf-8')   

v1 = json.loads(f1.read())      


for papers in v1:
    all_authors = []
    for authors in papers["authors"]:
        all_authors.append(authors["fname"])
    print(f'{papers["paper_id"]}:{all_authors}')

f1.close()   
```

```python
# Retrieve lname of authors of every publications + calculate number of authors
import json

f1 = open('dataset.json', encoding='utf-8')   

v1 = json.loads(f1.read())      


for papers in v1:
    all_authors = []
    i=0
    for authors in papers["authors"]:
        all_authors.append(authors["lname"])
        i=i+1
    print(f'{papers["paper_id"]}:{all_authors}')
    print(f'\tPaper id {papers["paper_id"]} has {i} authors.')

f1.close() 
```

```python
# Search papers according to author name
import json

f1 = open('dataset.json', encoding='utf-8')   

v1 = json.loads(f1.read())      

while True:
    flag = "Red"
    author_name = input ("Please input the name of author for the publication you are looking for: ")
    author_name = author_name.lower()
    author_name_validity = str(author_name).replace(" ","").replace(".","").replace("-", "")   # to avoid numbers and symbols, except for "-" and " " .

    if author_name_validity.isalpha() == False:   
        print("\nYou have entered an unexpected input, please retry.")   # error msg incase user input invalid
        flag = "Green"
    
    else:
        for papers in v1:
            for count in range (len(papers["authors"])):
                aufirst = papers['authors'][count]['fname']
                aulast = papers['authors'][count]['lname']
                if author_name in aufirst.lower() or author_name in aulast.lower():
                    flag = "Green"
                    print(f"\nName: {aufirst} {aulast}\nPublication: {papers['paper']['title']}")
                
    if flag == "Red":
        print(f"\nAuthor name containing '{author_name}' not found, please retype.")
    
    

f1.close()
```

```python
# Retrieve fullname of every author
import json

f1 = open("dataset.json", encoding = 'utf-8')

v1 = json.loads(f1.read())

for papers in v1:
    author_list = []
    for author in range (len(papers["authors"])):
        fullname = papers["authors"][author]["fname"] +" "+ papers["authors"][author]["lname"]
        author_list.append(fullname)
       
        print(f"Paper id:{papers['paper_id']} \nAuthor fullname: {author_list}")

        
f1.close()
```

```python
# Retrieve all keywords of every publication

import json

f1 = open("dataset.json", encoding = "utf-8")

v1 = json.loads(f1.read())

for publication in v1:
    keyword_list = []
    keyword_list.append(publication["paper"]["keywords"])

    for keywords in keyword_list:
        if keyword_list[0] != []:
            print(f"Paper id: {publication['paper_id']}\n\tKeywords: {keyword_list[0]}\n")
        else:
            print(f"Paper id: {publication['paper_id']}\n\tNo keyword found for this publication.\n")
       
```
