```python
#year

import json

f1 = open('master.json', encoding='utf-8')   

v1 = json.loads(f1.read())      

count = 0
for papers in v1:
    count = count + 1
    print(f"Paper id: {papers['paper_id']}\nPublication year: {papers['year']}\n")
print(count)

f1.close()   
```

```python
#author names

import json

f1 = open('master.json', encoding='utf-8')   

v1 = json.loads(f1.read())

paper_authors = {}

for paper in v1:
    paper_id = paper['paper_id']

    if paper_id not in paper_authors:
        paper_authors[paper_id] = []
    
    for author in paper['authors']:
        author_name = f"{author['fname']} {author['lname']}"
        paper_authors[paper_id].append(author_name)

for paper_id, authors in paper_authors.items():
    print(f"\nPaper ID: {paper_id}")
    print("Paper author(s):", ", ".join(authors))
    
f1.close()   
```

```python
#article title

import json

f1 = open('master.json', encoding='utf-8')   

v1 = json.loads(f1.read())

for paper in v1:
    print (f"Paper ID:{paper['paper_id']}\nTitle: {paper['paper']['title']}\n")

f1.close()   
```
