```python
# Very important! Please run this in the first place to get others to work
# Retrieval
import json

f1 = open('dataset.json', encoding='utf-8')   

v1 = json.loads(f1.read())  

from pprint import pprint

def limit_authorfullname():
    count = 0
    for papers in v1:
        allauthorlist = []
        for authors in papers["authors"]:
            fullname = f'{authors["fname"]} {authors["lname"]}'
            allauthorlist.append(fullname)
        print(f'Paper id {papers["paper_id"]}: {", ".join(allauthorlist)}') 

        count = count + 1
        if count >= 10:
            flag = "Red"
            while flag == "Red":

                view = input(f"\tType 'Y' to view more or type 'N' to stop: ")
                if view == "Y" or view == "y":
                    flag = "Y"
                    count = 0
                    break

                elif view == "N" or view == "n":
                    flag = "N"
                    break

                else:
                    print("You have entered an unexpected input.\n")

            if flag == "N":
                break
                return
            

def limit_numauthors_lname():
    count = 0
    for papers in v1:
        all_authors = []
        i=0
        for authors in papers["authors"]:
            all_authors.append(authors["lname"])
            i=i+1
        print(f'Paper id {papers["paper_id"]} has {i} authors. They are {", ".join(all_authors)}\n')
        count = count + 1
        if count >= 10:
            flag = "Red"
            while flag == "Red":

                view = input(f"\tType 'Y' to view more or type 'N' to stop: ")
                if view == "Y" or view == "y":
                    flag = "Y"
                    count = 0
                    break

                elif view == "N" or view == "n":
                    flag = "N"
                    break

                else:
                    print("You have entered an unexpected input.\n")

            if flag == "N":
                break
                return
                
def limit_keywords():
    count = 0
    for publication in v1:
        keyword_list = []
        keywords = ", ".join(publication["paper"]["keywords"])
        keyword_list.append(keywords)

        for keywords in keyword_list:
            if keyword_list[0] != '':
                print(f"Paper id: {publication['paper_id']}\n\tKeywords: {keyword_list[0]}\n")
            else:
                print(f"Paper id: {publication['paper_id']}\n\tNo keyword found for this publication.\n")
        count = count + 1
        if count >= 10:
            flag = "Red"
            while flag == "Red":

                view = input(f"\tType 'Y' to view more or type 'N' to stop: ")
                if view == "Y" or view == "y":
                    flag = "Y"
                    count = 0
                    break

                elif view == "N" or view == "n":
                    flag = "N"
                    break

                else:
                    print("You have entered an unexpected input.\n")

            if flag == "N":
                break
                return

def limit_cauthors():
    count = 0
    for papers in v1:
        found = False
        for no_allauthors in range(len(papers["authors"])):
            try:
                if papers["authors"][no_allauthors]["cauthor"] == True:
                    found = True
    #                 print(papers["authors"][no_allauthors]["cauthor"])
                    print(f"Paper id: {papers['paper_id']}\nContact author: {papers['authors'][no_allauthors]['fname']} {papers['authors'][no_allauthors]['lname']}\nContact email: {papers['authors'][no_allauthors]['email']}\n")
            except KeyError:
                flag = False
                print(f"Paper id: {papers['paper_id']}\nContact author: {papers['authors'][no_allauthors]['fname']} {papers['authors'][no_allauthors]['lname']}\nContact email: Email not found.\n")
        count = count + 1
        if count >= 10:
            flag = "Red"
            while flag == "Red":

                view = input(f"\tType 'Y' to view more or type 'N' to stop: ")
                if view == "Y" or view == "y":
                    flag = "Y"
                    count = 0
                    break

                elif view == "N" or view == "n":
                    flag = "N"
                    break

                else:
                    print("You have entered an unexpected input.\n")

            if flag == "N":
                break
                return

def limit_titles():
    count = 0
    for papers in v1:
        print(f"Paper id: {papers['paper_id']}\nPublication: {papers['paper']['title']}\n")
        count = count + 1
        if count >= 10:
            flag = "Red"
            while flag == "Red":

                view = input(f"\tType 'Y' to view more or type 'N' to stop: ")
                if view == "Y" or view == "y":
                    flag = "Y"
                    count = 0
                    break

                elif view == "N" or view == "n":
                    flag = "N"
                    break

                else:
                    print("You have entered an unexpected input.\n")

            if flag == "N":
                break
                return

def limit_printall():
    count = 0
    
    for papers in v1:
        allauthors = []
        keyword_list = []
        smalllistauthor = []
#         cauthorlist = []
        nameandcontactlist = []
        keywords = ", ".join(papers["paper"]["keywords"])
        keyword_list.append(keywords)
 
        for each_and_every_author in papers["authors"]:
            fullnames = f"{each_and_every_author['fname']}{each_and_every_author['lname']}"
            try: 
                 if each_and_every_author["cauthor"] == True:
                    tuple1 = (fullnames, each_and_every_author['email'])
#                     cauthorlist.append(tuple1)
                    string = f"{tuple1[0]}: {tuple1[1]}"
                    nameandcontactlist.append(string)
            except KeyError:
                nameandcontactlist.append("Email not provided.")

        
        for everyauthor in papers["authors"]:
            fullname = f"{everyauthor['fname']} {everyauthor['lname']}"
            smalllistauthor.append(fullname)
#         allauthors.append(", ".join(smalllistauthor))
        allauthors.append(smalllistauthor)

        
#         print(allauthors)
#         print(cauthorlist)
        for authors, keywords in zip(allauthors, keyword_list):                # zip(), is to make two loops to run simultaneously
            print(f"Paper id: {papers['paper_id']}")
            print(f"Authors: {', '.join(authors)}")
            print(f"Title: {papers['paper']['title']}")
            print(f"Year: {papers['year']}")
            print(f"Journal: {papers['source']['journal']}")
#             print(f"Contact authors and email: \n\t{(', '.join(map(str,nameandcontactlist))), sep = '\n'})")
            print('Contact authors and email: ')
            for contact_info in nameandcontactlist:
                print(f"\t{contact_info}")

    
            if keyword_list[0] != '':
                print(f"Keywords: {keyword_list[0]}")
            else:
                print(f"Keywords: No keyword found for this publication.")
                

            try:
                if papers['source']['volume'] == [] or papers['source']['volume'] == False:
                    print("Volume: Volume not found.")
                else:
                    print(f"Volume: {papers['source']['volume']}")
            except KeyError:
                print("Volume not provided.")

            try:
                if papers['source']['number'] == [] or papers['source']['number'] == False:
                    print("Number: Number not found.")
                else:
                    if type(papers['source']['number']) == list:
                        print(f"Number: {', '.join(map(str, papers['source']['number']))}")
                    else:
                        print(f"Number: {papers['source']['number']}")

            except KeyError:
                print("Number not provided.")

            print(f"Pages: {papers['source']['pages']['s_page']}-{papers['source']['pages']['e_page']}\n")
            count = count + 1
            flag = 'no flag'                              # to let last flag to work
            if count >= 10:
                while flag == 'no flag':
                    
                    view = input(f"\tType 'Y' to view more or type 'N' to stop: ")
                    if view == "Y" or view == "y":
                        flag = "Y"
                        count = 0
                        break

                    elif view == "N" or view == "n":
                        flag = "N"
                        break

                    else:
                        print("You have entered an unexpected input.\n")

                
        if flag == "N":
            break
            return

        
def this_is_retrieve_engine():
    import json
    f1 = open('dataset.json', encoding='utf-8')   

    v1 = json.loads(f1.read())      

    print('''\033[1m
    This is a program to retrieve info from the database.
    Please enter the corresponding number to start search:\033[0m''')
    print('''
        Name of authors:   1
        Number of authors: 2
        Keywords:          3
        Contact authors:   4
        Titles:            5
        All info:          6
        ''')
    search_term = input("Enter search term: ")


    if search_term == "1":
        limit_authorfullname()

    if search_term == "2":
        limit_numauthors_lname()

    if search_term == "3":
        limit_keywords()

    # consider adding number of keywords(?)

    if search_term == "4":
        limit_cauthors()

    if search_term == "5":
        limit_titles()

    if search_term == "6":
        limit_printall()
        
    else:
        print("You have entered an unexpected input, please restart and try again.")


f1.close()   

if __name__ == "__main__":           # just to "lock" the function
    this_is_retrieve_engine()
```
