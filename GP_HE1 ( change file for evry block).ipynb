```python
# Search for papers co-written by two authors
import json

f1 = open("dataset.json", encoding = "utf-8")

v1 = json.loads(f1.read())

while True:
    print("\033[1mThis program helps to search for papers that are co-written by multiple authors.\033[0m")


    author = input("Please enter name of  first author: ")
    authornext = input("Please enter name of next author: ")
    
    author1 = author.lower()
    authornext1 = authornext.lower()

    fullname_list = []
    author_list = []
    fulllist = []
    paperlist = []
    capital_list = []
    
    
    for papers in v1:
        small_list = []
        mini_capital_list = []
        for allauthors in range(len(papers["authors"])):
            
            capitalname = papers["authors"][allauthors]["fname"] + " " + papers["authors"][allauthors]["lname"]
            mini_capital_list.append(capitalname)
        
            fullname = (papers["authors"][allauthors]["fname"]).lower() + " " + (papers["authors"][allauthors]["lname"]).lower()
            small_list.append(fullname)
        fulllist.append(small_list)
        capital_list.append(mini_capital_list)

#     print(fulllist)

    for papers in v1:
        paperlist.append(papers["paper"]["title"])
#     print(paperlist)
    
        
    for authors in range(len(fulllist)):
        flag1 = "outside"
        flag2 = "outside"
        for i in range(len(fulllist[authors])):
            if author1 in fulllist[authors][i]:
                    flag1 = "inside"

            if authornext1 in fulllist[authors][i]:
                    flag2 = "inside"
      
        if flag1 == "inside" and flag2 == "inside":
            print(f"\nPaper ID: {authors+1}\nPublication: {paperlist[authors]}\nAuthors: {capital_list[authors]}\n")


f1.close()
```

```python
# Full search

import json

f1 = open("dataset.json", encoding="utf-8")
v1 = json.loads(f1.read())

papers_info_list = []

for papers in v1:
    smalllistauthor = []
    allkeyword = ", ".join(papers["paper"]["keywords"])
    for everyauthor in papers["authors"]:
        fullname = f"{everyauthor['fname']} {everyauthor['lname']}"
        smalllistauthor.append(fullname)
    authors = ", ".join(smalllistauthor)                            # unlike the previous one where we print all authors of every paper in a list, we can keep this as a variable, so we dont need to use a next list, we are still using v1
#     print(allkeyword)

    paper_info = {                                                  # since they are raw material obtain directly from papers, just create a dictionary with epected keys, this is a variable(which is a dictionary)
        "Paper id": papers['paper_id'],
        "Authors": authors,
        "Title": papers['paper']['title'],
        "Keywords": allkeyword,
        "Year": papers["year"],
        "Journal": papers["source"]["journal"]
    }
    
    try:
        if papers['source']['volume'] == []:                       # from the original json
            paper_info["Volume"] = "Volume not found."             # from the new dictinary variable created, append new key ["Volume"]
        else:
            paper_info["Volume"] = papers['source']['volume']
    except KeyError:
        paper_info["Volume"] = "Volume not provided."

    try:
        if papers['source']['number'] == []:
            paper_info["Number"] = "Number not found."             # append new key ["Number"]
        else:
            if type(papers['source']['number']) == list:
                paper_info["Number"] = ', '.join(map(str, papers['source']['number']))
            else:
                paper_info["Number"] = papers['source']['number']

    except KeyError:
        paper_info["Number"] = "Number not provided."
    
    paper_info["Pages"] = f"{papers['source']['pages']['s_page']}-{papers['source']['pages']['e_page']}"   # for the dictonary created, append new key ['pages'] and value

    papers_info_list.append(paper_info)           # append the whole paper_info to a list, so that we hv a database to search through

# pprint(papers_info_list)


search = input("Enter search term: ")


# Of course we can add search in other variable like year, publication... but i think more meaningful to do title and author!!

# output is the info of each dictionary
matching_papers = [info for info in papers_info_list if search.lower() in info["Title"].lower() or search.lower() in info["Authors"].lower()]

''' the info here is each and every dictionoary
# print every appropriate paper info into a list, as a compilation of all wanted papers
# same as: for info in papers_info_list if search.lower() in info["Title"].lower() or search.lower() in info["Authors"].lower():
# print(info)
'''

if matching_papers:                  # AKA if search has a result
    for paper in matching_papers:    # for info of wanted papers in the whole list of wanted papers
        print()
        print("\n".join([f"{key}: {value}" for key, value in paper.items()]))   # create tuple for each pair of key and value, define them, then join each pair with a new line
        print() 
else:
    print(f"No matching papers found for search term '{search}'.")

f1.close()
```
